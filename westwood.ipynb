```python
mydata = np.loadtxt("/content/drive/MyDrive/Colab Notebooks/3600/ch03/westwood.txt")
x = mydata[:,1:3]
y = mydata[:,3]
n,p=x.shape
```

```python
ixtx=np.linalg.inv( x.T @ x)
beta=ixtx @ x.T @ y
beta
```

```python
sse=y @ IH @ y
mse=sse/(n-p)
print(sse)
print(mse)
```

P.15

```python
varBeta=mse * ixtx
varBeta
```

```python
seBeta=np.sqrt(np.diagonal(varBeta))
seBeta
```

```python
myt=t(df=n-p).ppf(0.975)
myt
```

```python
CI=np.array([beta-myt*seBeta,beta+myt*seBeta])
CI
```
