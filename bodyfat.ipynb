```python
mydata = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/3600/bodyfat.csv")
mydata.head()
```

```python
x=mydata[['x1','x2','x3']].to_numpy()
n=x.shape[0]
x=np.c_[np.ones(n),x]
p=x.shape[1]
y=mydata['y']
```

```python
x
```

```python
sst=sum((y-y.mean())**2)
dft=n-1
print(f"sst= {sst:10.2f}")
```

$\hat{y}=\alpha + \beta_1 x_1 + \beta_2 x_2 + \beta_3 x_3$

```python
x123=x[:,[0,1,2,3]]
ixtx=np.linalg.inv(x123.T @ x123)
beta=ixtx @ x123.T @ y
print('beta=',beta)
yhat=x123 @ beta
res=y-yhat

sse123=np.sum(res**2)
dfe123=n-4
mse123=sse123/dfe123

ss123=sst-sse123
df123=3
ms123=ss123/df123

print(f"sse={sse123:10.2f}, ss123={ss123:10.2f}")
```

$\hat{y}=\alpha + \beta_1 x_1 $

```python
x1=x[:,[0,1]]
ixtx1=np.linalg.inv(x1.T @ x1)
beta1=ixtx1 @ x1.T @ y
print('beta=',beta1)

yhat=x1 @ beta1
res=y-yhat

sse1=np.sum(res**2)
dfe1=n-2

ssx1=sst-sse1
dfx1=1
msx1=ssx1/dfx1

fx1=msx1/mse123
ftest=msx1/sse1*dfe1

print(f"sse={sse1:10.2f}, ssx1={ssx1:10.2f}")
```

$\hat{y}=\alpha + \beta_1 x_1 + \beta_2 x_2 $

```python
x12=x[:,[0,1,2]]
ixtx12=np.linalg.inv(x12.T @ x12)
beta12=ixtx12 @ x12.T @ y
print('beta=',beta12)

yhat=x12 @ beta12
res=y-yhat

sse12=np.sum(res**2)
dfe12=n-2
ssx12=sst-sse12
dfx12=1
print(f"sse={sse12:10.2f}")
```

Type 1 (sequential) sum of squares

```python
ss2_1=sse1-sse12
ms2_1=ss2_1
ss3_12=sse12-sse123
ms3_12=ss3_12/1

print (f"{'source':<12} {'ss':>12} {'df':>3} {'ms':>12}")
print (f"{'X1':<12} {ssx1:12.2f} {1:>3} {msx1:12.2f}")
print (f"{'X2|X1':<12} {ss2_1:12.2f} {1:>3} {ms2_1:12.2f}")
print (f"{'X3|X1X2':<12} {ss3_12:12.2f} {1:>3} {ms3_12:12.2f}")
print (f"{'error':<12} {sse:12.2f} {dfe:>3} {mse:12.2f}")
print (f"{'total':<12} {sst:12.2f} {dft:>3}")
```

$\hat{y}=\alpha + \beta_2 x_2 + \beta_2 x_2 $

```python
x23=x[:,[0,2,3]]
ixtx23=np.linalg.inv(x23.T @ x23)
beta23=ixtx23 @ x23.T @ y
print("beta=",beta23)

yhat=x23 @ beta23
res=y-yhat

sse23=np.sum(res**2)
dfe23=n-2
ssx23=sst-sse23
dfx23=1

print(f"sse= {sse23:10.2f}")
```

$\hat{y}=\alpha + \beta_1 x_1 + \beta_3 x_3 $

```python
x13=x[:,[0,1,3]]
ixtx13=np.linalg.inv(x13.T @ x13)
beta13=ixtx13 @ x13.T @ y
print("beta=",beta13)

yhat=x13 @ beta13
res=y-yhat
sse13=np.sum(res**2)
dfe13=n-2
mse13=sse13/dfe13
ssx13=sse13-sse
print(f"sse={sse13:10.2f}")
```

```python
ss1_23=sse23-sse123
ms1_23=ss1_23/1

ss2_13=sse13-sse123
ms2_13=ss2_13/1

f1_23=ms1_23/mse123
f2_13=ms2_13/mse123
f3_12=ms3_12/mse123
f123=ms123/mse123

print (f"{'source':<12} {'ss':>12} {'df':>3} {'ms':>12} {'F':>12}")
print (f"{'X1X2X3':<12} {ss123:12.2f} {3:>3} {ms123:12.2f} {f123:12.2f} ")
print (f"{'X1|X2X3':<12} {ss1_23:12.2f} {1:>3} {ms1_23:12.2f} {f1_23:12.2f} ")
print (f"{'X2|X1X3':<12} {ss2_13:12.2f} {1:>3} {ms2_13:12.2f} {f2_13:12.2f}")
print (f"{'X3|X1X2':<12} {ss3_12:12.2f} {1:>3} {ms3_12:12.2f} {f3_12:12.2f}")
print (f"{'error':<12} {sse:12.2f} {dfe:>3} {mse:12.2f}")
print (f"{'total':<12} {sst:12.2f} {dft:>3}")
```
